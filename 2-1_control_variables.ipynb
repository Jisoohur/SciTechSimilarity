# **논문 관련 통제변수**

**논문 기관 수**

```python
import pandas as pd
pub_institution = pd.read_csv('eu_pub_prox_institution.csv')
pub_institution.head()
```

```python
# create 'period' column 
def assign_period(pubyear):
    if 1981 <= pubyear <= 1985:
        return 1
    elif 1986 <= pubyear <= 1990:
        return 2 
    elif 1991 <= pubyear <= 1995:
        return 3 
    elif 1996 <= pubyear <= 2000:
        return 4
    elif 2001 <= pubyear <= 2005:
        return 5
    elif 2006 <= pubyear <= 2010:
        return 6
    elif 2011 <= pubyear <= 2015:
        return 7
    elif 2016 <= pubyear <= 2020:
        return 8
    else:
        return 0 
    
pub_institution['period'] = pub_institution['pubyear'].apply(assign_period)
pub_institution.head()
```

```python
# slice eu_nuts_id to 4 digits
pub_institution['eu_nuts_id'] = pub_institution['eu_nuts_id'].str.slice(0, 4)
pub_institution.head()
```

```python
# group by 'eu_nuts_id' and 'period', then count unique 'institution_id_te' 
num_institution = pub_institution.groupby(['eu_nuts_id', 'period'])['institution_id_te'].nunique().reset_index()
num_institution.rename(columns={'institution_id_te': 'num_institution'}, inplace=True)
num_institution.head()
```

```python
pub_subject = pd.merge(pub_subject, pub_institution, on='pubid', how='left')
pub_subject.head()
```

```python
# create 'period' column 
def assign_period(pubyear):
    if 1981 <= pubyear <= 1985:
        return 1
    elif 1986 <= pubyear <= 1990:
        return 2 
    elif 1991 <= pubyear <= 1995:
        return 3 
    elif 1996 <= pubyear <= 2000:
        return 4
    elif 2001 <= pubyear <= 2005:
        return 5
    elif 2006 <= pubyear <= 2010:
        return 6
    elif 2011 <= pubyear <= 2015:
        return 7
    elif 2016 <= pubyear <= 2020:
        return 8
    else:
        return 0 
    
pub_subject['period'] = pub_subject['pubyear'].apply(assign_period)
pub_subject.head()
```

**funding 수**

```python
pub_funding = pd.read_csv('eu_pub_prox_funding.csv')
pub_funding
```

```python
pub_funding = pd.merge(pub_funding, pub_institution, on='pubid', how='left')
pub_funding.head()
```

```python
# create 'period' column 
def assign_period(pubyear):
    if 1981 <= pubyear <= 1985:
        return 1
    elif 1986 <= pubyear <= 1990:
        return 2 
    elif 1991 <= pubyear <= 1995:
        return 3 
    elif 1996 <= pubyear <= 2000:
        return 4
    elif 2001 <= pubyear <= 2005:
        return 5
    elif 2006 <= pubyear <= 2010:
        return 6
    elif 2011 <= pubyear <= 2015:
        return 7
    elif 2016 <= pubyear <= 2020:
        return 8
    else:
        return 0 
    
pub_funding['period'] = pub_funding['pubyear'].apply(assign_period)
pub_funding.head()
```

```python
# group by 'eu_nuts_id' and 'period', then count unique 'grant_id' 
pub_funding_1 = pub_funding.groupby(['eu_nuts_id', 'period'])['grant_id'].nunique().reset_index()
pub_funding_1.rename(columns={'grant_id': 'num_funding'}, inplace=True)
pub_funding_1.head()
```

```python
# group by 'period' and 'eu_nuts_id', then count unique 'pubid' 
pub_funding_2 = pub_funding.groupby(['eu_nuts_id', 'period'])['pubid'].nunique().reset_index()
pub_funding_2.rename(columns={'pubid': 'num_pubs'}, inplace=True)
pub_funding_1 = pd.merge(pub_funding_1, pub_funding_2, on=['eu_nuts_id', 'period'], how='left')
pub_funding_1.head()
```

```python
# calculate the ratio of num_funding to num_pubs
pub_funding_1['funding_pub_ratio'] = pub_funding_1['num_funding'] / pub_funding_1['num_pubs']
pub_funding_1.head()

# drop num_pubs column
pub_funding_1.drop(columns=['num_pubs'], inplace=True)
pub_funding_1.head()
```

```python
# group by 'eu_nuts_id' and 'period' and count number of unique 'agency' 
pub_funding_2 = pub_funding.groupby(['eu_nuts_id', 'period'])['agency'].nunique().reset_index()
pub_funding_2.rename(columns={'agency': 'num_funding_agencies'}, inplace=True)
pub_funding_2.head()
```

```python
# merge the two funding dataframes
pub_funding = pd.merge(pub_funding_1, pub_funding_2, on=['eu_nuts_id', 'period'], how='left')
pub_funding.head()
```

**모든 논문 관련 통제변수 merge**

```python
pub_control = pd.merge(num_institution, pub_author, on=['eu_nuts_id', 'period'], how='outer')
pub_control = pd.merge(pub_control, pub_subject, on=['eu_nuts_id', 'period'], how='outer')
pub_control = pd.merge(pub_control, pub_funding, on=['eu_nuts_id', 'period'], how='outer')
pub_control.head()
```

```python
# include pub in every column name 
pub_control.columns = ['eu_nuts_id', 'period', 
                       'pub_num_institution', 
                       'pub_num_authors', 
                       'pub_num_subjects', 
                       'pub_num_funding', 
                       'pub_funding_ratio', 
                       'pub_num_funded_agencies']
pub_control.head()
```

```python
pub_control.to_csv('2-1_pub_control.csv')
```

# **특허 관련 통제변수**

**발명가 수**

```python
patent_pers_1 = pd.read_csv('patent_pers_1.csv')
patent_pers_2 = pd.read_csv('patent_pers_2.csv')
patent_pers_3 = pd.read_csv('patent_pers_3.csv')

patent_pers = pd.concat([patent_pers_1, patent_pers_2, patent_pers_3], ignore_index=True)
patent_pers
```

```python
# drop rows where nuts_level is 0 
patent_pers = patent_pers[patent_pers['nuts_level'] != 0]
patent_pers 
```

```python
# drop rows where 'nuts' is NaN
patent_pers = patent_pers.dropna(subset=['nuts'])
patent_pers
```

```python
patent_pers['nuts_level'].value_counts()
```

```python
# remove rows where nuts_level is 1 or 2 
patent_pers = patent_pers[~patent_pers['nuts_level'].isin([1, 2])]
patent_pers
```

```python
patent_pers = patent_pers[patent_pers['nuts'] != 'AN000']

# slice nuts to 4 digits
patent_pers['nuts'] = patent_pers['nuts'].str.slice(0, 4)
patent_pers
```

```python
# change nuts to eu_nuts_id 
patent_pers.rename(columns={'nuts': 'eu_nuts_id'}, inplace=True)
patent_pers
```

```python
patent_app = pd.read_csv('patent_app.csv')
patent_app = patent_app[['appln_id', 'appln_filing_year']]
patent_app.rename(columns={'appln_filing_year': 'pubyear'}, inplace=True)
patent_app
```

```python
# merge 
patent_pers = pd.merge(patent_pers, patent_app, on='appln_id', how='left')
patent_pers.head()
```

```python
# create 'period' column 
def assign_period(pubyear):
    if 1981 <= pubyear <= 1985:
        return 1
    elif 1986 <= pubyear <= 1990:
        return 2 
    elif 1991 <= pubyear <= 1995:
        return 3 
    elif 1996 <= pubyear <= 2000:
        return 4
    elif 2001 <= pubyear <= 2005:
        return 5
    elif 2006 <= pubyear <= 2010:
        return 6
    elif 2011 <= pubyear <= 2015:
        return 7
    elif 2016 <= pubyear <= 2020:
        return 8
    else:
        return 0 

patent_pers['period'] = patent_pers['pubyear'].apply(assign_period)
patent_pers.head()
```

```python
patent_pers['psn_sector'].value_counts()
```

```python
# group by 'eu_nuts_id' and 'period', then count unique 'person_id' where psn_sector is not 
# 'COMPANY', 'GOV NON-PROFIT', 'UNIVERSITY', 'HOSPITAL', 'COMPANY GOV NON-PROFIT', 'COMPANY HOSPITAL',
# 'COMPANY UNIVERSITY', 'GOV NON-PROFIT UNIVERSITY', 'COMPANY GOV NON-PROFIT UNIVERSITY', 'GOV NON-PROFIT HOSPITAL' 

num_inventor = patent_pers[~patent_pers['psn_sector'].isin(['COMPANY', 'GOV NON-PROFIT', 'UNIVERSITY',
                                                       'HOSPITAL', 'COMPANY GOV NON-PROFIT', 
                                                       'COMPANY HOSPITAL', 'COMPANY UNIVERSITY', 
                                                       'GOV NON-PROFIT UNIVERSITY', 
                                                       'COMPANY GOV NON-PROFIT UNIVERSITY', 
                                                       'GOV NON-PROFIT HOSPITAL'])].groupby(
    ['eu_nuts_id', 'period'])['person_id'].nunique().reset_index()
num_inventor.rename(columns={'person_id': 'num_inventors'}, inplace=True)
num_inventor.head()
```

```python
# group by 'eu_nuts_id' and 'period', then count unique 'person_id' where psn_sector is not 'INDIVIDUAL', 'UNKNOWN' 
num_organization = patent_pers[~patent_pers['psn_sector'].isin(['INDIVIDUAL', 'UNKNOWN'])].groupby(
    ['eu_nuts_id', 'period'])['person_id'].nunique().reset_index()
num_organization.rename(columns={'person_id': 'num_organizations'}, inplace=True)
num_organization.head()
```

```python
# group by 'eu_nuts_id' and 'period', then count unique 'person_id' where psn_sector is 'COMPANY' 
num_company = patent_pers[patent_pers['psn_sector'] == 'COMPANY'].groupby(
    ['eu_nuts_id', 'period'])['person_id'].nunique().reset_index() 
num_company.rename(columns={'person_id': 'num_companies'}, inplace=True)
num_company.head()
```

```python
patent_pers['psn_sector'].value_counts()
```

```python
# group by 'eu_nuts_id' and 'period', then count unique 'person_id' where psn_sector is 'UNIVERSITY', 'COMPANY UNIVERSITY', 'GOV NON-PROFIT UNIVERSITY', 'COMPANY GOV NON-PROFIT UNIVERSITY' 
num_university = patent_pers[patent_pers['psn_sector'].isin(['UNIVERSITY', 'COMPANY UNIVERSITY',
                                                       'GOV NON-PROFIT UNIVERSITY', 
                                                       'COMPANY GOV NON-PROFIT UNIVERSITY'])].groupby(
    ['eu_nuts_id', 'period'])['person_id'].nunique().reset_index()
num_university.rename(columns={'person_id': 'num_universities'}, inplace=True)
num_university.head()
```

```python
# group by 'eu_nuts_id' and 'period', then count unique 'person_id' where psn_sector is 'GOV NON-PROFIT' 
num_gov_institution = patent_pers[patent_pers['psn_sector'] == 'GOV NON-PROFIT'].groupby(
    ['eu_nuts_id', 'period'])['person_id'].nunique().reset_index()
num_gov_institution.rename(columns={'person_id': 'num_gov_institutions'}, inplace=True)
num_gov_institution.head()
```

```python
# merge all the dataframes
num_authors = pd.merge(num_inventor, num_organization, on=['eu_nuts_id', 'period'], how='outer')
num_authors = pd.merge(num_authors, num_company, on=['eu_nuts_id', 'period'], how='outer')
num_authors = pd.merge(num_authors, num_university, on=['eu_nuts_id', 'period'], how='outer')
num_authors = pd.merge(num_authors, num_gov_institution, on=['eu_nuts_id', 'period'], how='outer')
num_authors.head()
```

```python
# fill NaN values with 0
num_authors.fillna(0, inplace=True)
num_authors.head()
```

```python
patent_cpc = pd.read_csv('patent_cpc.csv')
patent_cpc
```

```python
# slice cpc_class_symbol to the first 4 characters
patent_cpc['cpc_4digit'] = patent_cpc['cpc_class_symbol'].str.slice(0, 4)
patent_cpc
```

```python
# create a new column 'cpc_8digit'
patent_cpc['cpc_8digit'] = patent_cpc['cpc_class_symbol'].str.extract(r'^([^/]+)')[0]
patent_cpc
```

```python
patent_app = pd.read_csv('patent_app.csv')
patent_app = patent_app[['appln_id', 'appln_filing_year']]
patent_app.rename(columns={'appln_filing_year': 'pubyear'}, inplace=True)
patent_app
```

```python
patent_cpc = pd.merge(patent_cpc, patent_app, on='appln_id', how='left')
patent_cpc.head()
```

```python
# create 'period' column 
def assign_period(pubyear):
    if 1981 <= pubyear <= 1985:
        return 1
    elif 1986 <= pubyear <= 1990:
        return 2 
    elif 1991 <= pubyear <= 1995:
        return 3 
    elif 1996 <= pubyear <= 2000:
        return 4
    elif 2001 <= pubyear <= 2005:
        return 5
    elif 2006 <= pubyear <= 2010:
        return 6
    elif 2011 <= pubyear <= 2015:
        return 7
    elif 2016 <= pubyear <= 2020:
        return 8
    else:
        return 0 

patent_cpc['period'] = patent_cpc['pubyear'].apply(assign_period)
patent_cpc.head()
```

```python
patent_pers_1 = pd.read_csv('patent_pers_1.csv')
patent_pers_2 = pd.read_csv('patent_pers_2.csv')
patent_pers_3 = pd.read_csv('patent_pers_3.csv')

patent_pers = pd.concat([patent_pers_1, patent_pers_2, patent_pers_3], ignore_index=True)
patent_pers = patent_pers[['appln_id', 'nuts', 'nuts_level']]  # correct column selection
patent_pers = patent_pers.drop_duplicates()
patent_pers
```

```python
# drop rows where nuts_level is 0 
patent_pers = patent_pers[patent_pers['nuts_level'] != 0]

# drop rows where 'nuts' is NaN
patent_pers = patent_pers.dropna(subset=['nuts'])

# remove rows where nuts_level is 1 or 2 
patent_pers = patent_pers[~patent_pers['nuts_level'].isin([1, 2])]

patent_pers = patent_pers[patent_pers['nuts'] != 'AN000']

# slice nuts to 4 digits
patent_pers['nuts'] = patent_pers['nuts'].str.slice(0, 4)

# change nuts to eu_nuts_id 
patent_pers.rename(columns={'nuts': 'eu_nuts_id'}, inplace=True)
patent_pers
```

```python
# drop nuts_level column
patent_pers.drop(columns=['nuts_level'], inplace=True)
patent_pers
```

```python
patent_cpc = pd.merge(patent_cpc, patent_pers, on='appln_id', how='left')
patent_cpc.head()
```

```python
# group by 'eu_nuts_id' and 'period', then count unique 'cpc_4digit'
patent_cpc_4digit = patent_cpc.groupby(['eu_nuts_id', 'period'])['cpc_4digit'].nunique().reset_index()
patent_cpc_4digit.rename(columns={'cpc_4digit': 'num_cpc_4digit'}, inplace=True)
patent_cpc_4digit.head()
```

```python
# group by 'eu_nuts_id' and 'period', then count unique 'cpc_8digit' 
patent_cpc_8digit = patent_cpc.groupby(['eu_nuts_id', 'period'])['cpc_8digit'].nunique().reset_index()
patent_cpc_8digit.rename(columns={'cpc_8digit': 'num_cpc_8digit'}, inplace=True)
patent_cpc_8digit.head()
```

```python
patent_cpc = pd.merge(patent_cpc_4digit, patent_cpc_8digit, on=['eu_nuts_id', 'period'], how='outer')
patent_cpc.head()
```

**특허 관련 통제변수 merge**

```python
patent_control = pd.merge(num_authors, patent_cpc, on=['eu_nuts_id', 'period'], how='outer')
patent_control.head()
```
